```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
filepath = '/home/hansm/active_learning/Double_pendulum/GP/metrics.csv'
df = pd.read_csv(filepath)
filepath2 = '/home/hansm/active_learning/Double_pendulum/GPytorch/output_data/metrics.csv'
df2 = pd.read_csv(filepath2)
```

```python
rmse_df = df['RMSE']
rmse = rmse_df.values
rmse_df2 = df2['RMSE']
rmse2 = rmse_df2.values
```

```python
print(len(rmse))
print(len(rmse2))
```

```python
print(rmse[-1])
```

```python
plt.plot(rmse, label='scikit-learn')
plt.plot(rmse2, label='GPytorch with added jitter')
plt.legend()
plt.xlabel("N iterations")
plt.ylabel("Test RMSE")
plt.title("Double pendulum Gpytorch with added jitter vs scikit-learn")
plt.show()
```
